```python
dataset = pd.read_csv("kc_house_data.csv")
dataset = dataset.loc[0:200,"price":"floors"]
dataset
```

```python
dataset = (dataset-dataset.mean())/(dataset.min()-dataset.max())
```

```python
dataset
```

```python
y_output = dataset["price"]
y_output
```

```python
x_input = dataset.loc[:,"bedrooms":]
x_input.insert(0,"ones",1)
x_input
```

```python
theta =   np.linalg.inv(x_input.T @ x_input) @ x_input.T @ y_output
theta
```

```python
h = np.dot(x_input,theta)
h.sort()
```

```python
plt.scatter(x_input['sqft_living'],y_output, color='purple')
plt.xlabel('sqft_living (square feet)')
plt.ylabel('Price')
plt.title('Relationship between House Size and Price')
plt.plot(x_input['sqft_living'],h, color='red')

plt.show()
```
