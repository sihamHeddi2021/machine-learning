```python
x_input = des.loc[:,"Gender":"Food and drink"]
x_input
```

```python
y_output = des.loc[:,"satisfaction"]
y_output.value_counts()
```

```python
x_input.insert(0,"ones",1)
x_input
```

```python
n = x_input.shape[1] # number of features
m = x_input.shape[0] # number of rows
```

```python
def hypothesis(theta, X):
    z = np.dot(X, theta).astype(float)
    h = 1 / (1 + np.exp(-z))
    return h
```

```python

def gradient_descent(alpha,steps):
    theta = np.ones(n+1)
    theta[0] = 0.3
    
    for i in range(steps):
        h = hypothesis(theta,x_input) 
        gradient = np.dot(x_input.T, (h - y_output)) / m
        theta = theta - alpha * gradient
    return theta
theta=gradient_descent(0.02,400)
    
```
